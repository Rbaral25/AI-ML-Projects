# Clean review_body

````python
# Clean review_body
def clean_text(text):
    text = re.sub(r'<.*?>', '', str(text))  # Remove HTML
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)  # Remove special chars
    return text.lower().strip()

df['review_body'] = df['review_body'].apply(clean_text)
````

# Hypothesis testing: Verified vs non-verified ratings

````python
# Hypothesis testing: Verified vs non-verified ratings
verified_ratings = df[df['verified_purchase'] == 'Y']['star_rating']
non_verified = df[df['verified_purchase'] == 'N']['star_rating']
t_stat, p_val = stats.ttest_ind(verified_ratings, non_verified)
print(f"T-test: T-stat={t_stat}, P-value={p_val}")
````

# Bayesian probability: P(Helpful | Rating > 3)

````python
# Bayesian probability: P(Helpful | Rating > 3)
prior_helpful = (df['helpful_votes'] > 0).mean()
likelihood = ((df['helpful_votes'] > 0) & (df['star_rating'] > 3)).sum() / (df['star_rating'] > 3).sum()
posterior = (likelihood * prior_helpful) / (df['star_rating'] > 3).mean()
print(f"P(Helpful | High Rating): {posterior}")
````

# PyTorch DNN

````python
# PyTorch DNN
class DNN(nn.Module):
    def __init__(self, input_size):
        super().__init__()
        self.fc1 = nn.Linear(input_size, 128)
        self.fc2 = nn.Linear(128, 64)
        self.fc3 = nn.Linear(64, 1)

    def forward(self, x):
        x = torch.relu(self.fc1(x))
        x = torch.relu(self.fc2(x))
        return self.fc3(x)

X_train_t = torch.tensor(X_train.values.astype(float), dtype=torch.float32)
y_train_t = torch.tensor(y_train.values.astype(float), dtype=torch.float32).unsqueeze(1)
X_test_t = torch.tensor(X_test.values.astype(float), dtype=torch.float32)
y_test_t = torch.tensor(y_test.values.astype(float), dtype=torch.float32).unsqueeze(1)

model_pt = DNN(X_train.shape[1])
criterion = nn.MSELoss()
optimizer = optim.Adam(model_pt.parameters(), lr=0.001)
for epoch in range(10):
    optimizer.zero_grad()
    output = model_pt(X_train_t)
    loss = criterion(output, y_train_t)
    loss.backward()
    optimizer.step()
print("PyTorch DNN - Loss:", criterion(model_pt(X_test_t), y_test_t).item())

````
