# Hypothesis testing: Verified vs non-verified ratings

```python
# Hypothesis testing: Verified vs non-verified ratings
verified_ratings = df[df['verified_purchase'] == 'Y']['star_rating']
non_verified = df[df['verified_purchase'] == 'N']['star_rating']
t_stat, p_val = stats.ttest_ind(verified_ratings, non_verified)
print(f"T-test: T-stat={t_stat}, P-value={p_val}")
```

Observation -
Because the P-value (3.029e-12) is much smaller than a typical significance level (e.g., 0.05), we have strong statistical evidence to reject the null hypothesis.</br>
This means that there is a statistically significant difference between the mean star ratings of verified purchases and non-verified purchases in this dataset. While the T-statistic suggests the verified purchase ratings are slightly lower on average, the primary finding from the hypothesis test is that the difference observed is unlikely to be due to random chance.


# Bayesian probability: P(Helpful | Rating > 3)

```python
# Bayesian probability: P(Helpful | Rating > 3)
prior_helpful = (df['helpful_votes'] > 0).mean()
likelihood = ((df['helpful_votes'] > 0) & (df['star_rating'] > 3)).sum() / (df['star_rating'] > 3).sum()
posterior = (likelihood * prior_helpful) / (df['star_rating'] > 3).mean()
print(f"P(Helpful | High Rating): {posterior}")
```

Observation-
The Bayesian probability calculation shows that even among reviews with high star ratings (greater than 3), the likelihood of a review being marked as helpful is relatively low, at about 8.5%. This aligns with the earlier observation from the summary statistics that while ratings are generally high, helpful votes are sparse for most reviews.

This analysis suggests that a high star rating does not strongly predict whether a review will be considered helpful by other users in this dataset. There might be other factors influencing helpfulness votes besides just the rating itself.

```python


# Install required packages (run in shell / notebook)

!pip install ultralytics  # YOLOv8 (PyTorch) training+export helpers
!pip install opencv-python-headless pillow tqdm pycocotools lxml
# Optional: labelImg for manual annotation GUI (outside notebook)
!pip install labelImg

# ----------------------------
# Yolo - Expected dataset layout (YOLO format)
# ----------------------------
# dataset/
#   ├── images/
#   │    ├── train/
#   │    │    ├── img_001.jpg
#   │    │    ├── img_002.jpg
#   │    │    └── ...
#   │    ├── val/
#   │    └── test/
#   └── labels/
#        ├── train/
#        │    ├── img_001.txt
#        │    ├── img_002.txt
#        │    └── ...
#        ├── val/
#        └── test/
#
# Each .txt file (YOLO format) contains one line per object:
# <class_id> <x_center_norm> <y_center_norm> <width_norm> <height_norm>
#
# Example data.yaml for YOLOv8:
# train: /path/to/dataset/images/train
# val: /path/to/dataset/images/val
# test: /path/to/dataset/images/test
# nc: 3
# names: ['apparel','electronics','home_kitchen']


# Utility: convert bounding boxes (COCO or Pascal VOC) -> YOLO txt

import os
import json
import xml.etree.ElementTree as ET
from pathlib import Path
from PIL import Image
import numpy as np

def voc_xml_to_yolo(txt_out_dir, xml_path, class_map):
    """
    Convert one Pascal VOC .xml annotation to YOLO .txt.
    class_map: dict mapping class name -> class_id, e.g. {'Apparel':0, 'Electronics':1}
    Writes a .txt file with the same base name in txt_out_dir.
    """
    tree = ET.parse(xml_path)
    root = tree.getroot()
    size = root.find('size')
    w = float(size.find('width').text)
    h = float(size.find('height').text)
    lines = []
    for obj in root.findall('object'):
        cls_name = obj.find('name').text
        if cls_name not in class_map:
            continue
        cls_id = class_map[cls_name]
        bbox = obj.find('bndbox')
        xmin = float(bbox.find('xmin').text)
        ymin = float(bbox.find('ymin').text)
        xmax = float(bbox.find('xmax').text)
        ymax = float(bbox.find('ymax').text)
        # convert to YOLO format (normalized center x,y,w,h)
        x_c = ((xmin + xmax) / 2.0) / w
        y_c = ((ymin + ymax) / 2.0) / h
        bw = (xmax - xmin) / w
        bh = (ymax - ymin) / h
        lines.append(f"{cls_id} {x_c:.6f} {y_c:.6f} {bw:.6f} {bh:.6f}")
    out_file = os.path.join(txt_out_dir, Path(xml_path).stem + ".txt")
    with open(out_file, "w") as f:
        f.write("\n".join(lines))

def coco_json_to_yolo(txt_out_dir, coco_json_path, images_dir, class_map):
    """
    Convert COCO json annotations to YOLO .txt files.
    class_map maps COCO category names -> class ids (or map category_id -> class_id)
    """
    with open(coco_json_path, "r") as f:
        coco = json.load(f)
    imgs = {im['id']: im for im in coco['images']}
    # map category_id -> desired class_id (0..nc-1)
    cat_to_cls = {}
    for cat in coco['categories']:
        name = cat['name']
        if name in class_map:
            cat_to_cls[cat['id']] = class_map[name]
    anns_per_img = {}
    for ann in coco['annotations']:
        img_id = ann['image_id']
        anns_per_img.setdefault(img_id, []).append(ann)
    for img_id, ann_list in anns_per_img.items():
        im = imgs[img_id]
        iw, ih = im['width'], im['height']
        lines = []
        for ann in ann_list:
            catid = ann['category_id']
            if catid not in cat_to_cls:
                continue
            cls_id = cat_to_cls[catid]
            x, y, bw, bh = ann['bbox']  # COCO bbox: [x_min, y_min, width, height]
            x_c = (x + bw/2.0) / iw
            y_c = (y + bh/2.0) / ih
            lines.append(f"{cls_id} {x_c:.6f} {y_c:.6f} {bw/iw:.6f} {bh/ih:.6f}")
        out_file = os.path.join(txt_out_dir, im['file_name'].rsplit(".",1)[0] + ".txt")
        with open(out_file, "w") as f:
            f.write("\n".join(lines))


# Build data.yaml programmatically for YOLOv8

# Update data.yaml with the correct paths to the split image directories
split_base_dir = "/content/Images-final-capstone_split"
train_dir = os.path.join(split_base_dir, "train")
validation_dir = os.path.join(split_base_dir, "validation")
test_dir = os.path.join(split_base_dir, "test")


data_yaml = {
    'train': train_dir,  # Use the correct training directory path
    'val': validation_dir, # Use the correct validation directory path
    'test': test_dir, # Use the correct test directory path
    'nc': 1, # Update class count based on previous observation (only 'Apparel' found)
    'names': ['Apparel'] # Update class names accordingly
}
# Save YAML
import yaml
with open("data.yaml", "w") as f:
    yaml.dump(data_yaml, f)
print("Wrote data.yaml ->", os.path.abspath("data.yaml"))


from ultralytics import YOLO

# create a model (you can start from pretrained weights 'yolov8n.pt')
# If you have a custom model config, put that path here
model = YOLO("yolov8n.pt")  # small/presized backbone

# Train: specify data config file and hyperparameters
# This will create runs/train/... with logs and weights
# Replace 'data.yaml' path with your data.yaml created above.
# Note: We still need corresponding label files in the same structure as images.
# The current unzipping/splitting only handled images.
# For actual object detection training, you need bounding box annotations (.txt files)
# in the 'labels/train', 'labels/val', 'labels/test' directories.
# If the zip file didn't contain these, they need to be generated (e.g., via manual annotation).
# Since we don't have the label files from the provided zip, the training will likely fail
# when it looks for the label files corresponding to the images.
# I will update the data.yaml paths, but note that label files are required for training.

model.train(data="data.yaml", epochs=50, imgsz=640, batch=16, workers=4)
# If out of memory, lower imgsz and/or batch, or use yolov8n.


# Convert a TensorFlow SavedModel -> TFLite (if you have a SavedModel)
import tensorflow as tf

saved_model_dir = "/content/Images-final-capstone_split"  # replace with your savedmodel dir @TODO - incorrect
if os.path.exists(saved_model_dir):
    converter = tf.lite.TFLiteConverter.from_saved_model(saved_model_dir)
    # Optimization options - use if quantization is desired (smaller / faster model)
    converter.optimizations = [tf.lite.Optimize.DEFAULT]
    # For full integer quantization, provide a representative dataset function:
    # converter.representative_dataset = representative_data_gen
    # converter.target_spec.supported_ops = [tf.lite.OpsSet.TFLITE_BUILTINS_INT8]
    tflite_model = converter.convert()
    with open("model.tflite", "wb") as f:
        f.write(tflite_model)
    print("Saved TFLite model to model.tflite")
else:
    # print("SavedModel not found at", saved_model_dir)
    pass # Avoid printing this message unless explicitly relevant

```

# Retrieval-Augmented Generation (RAG) implementation


```python
# Retrieval-Augmented Generation (RAG) Using Hugging Face models — no external API keys required

# Install dependencies (run only once)
# !pip install transformers sentence-transformers faiss-cpu datasets torch --quiet
# !pip install langchain-community --quiet # For LangChain components
# !pip install rank_bm25 --quiet # For BM25 retriever


from sentence_transformers import SentenceTransformer
from datasets import load_dataset
import faiss
import numpy as np


# Load a small dataset for retrieval  from the Hugging Face datasets library.
# Use a small subset of the 'ag_news' dataset for demo
dataset = load_dataset("ag_news", split="train[:200]")
documents = dataset['text']

print(f"Loaded {len(documents)} documents for retrieval.")

# 3️Generate embeddings for documents

embedder = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2")
doc_embeddings = embedder.encode(documents, convert_to_numpy=True, show_progress_bar=True)


# Create a FAISS vector store for retrieval
dim = doc_embeddings.shape[1]
index = faiss.IndexFlatL2(dim)
index.add(doc_embeddings)
print(f"FAISS index contains {index.ntotal} documents.")

#  Define the retriever function

def retrieve_docs(query, top_k=3):
    query_vec = embedder.encode([query])
    distances, indices = index.search(query_vec, top_k)
    # Convert numpy indices to a list of integers before indexing the dataset object
    retrieved_texts = [documents[int(i)] for i in indices[0]]
    return retrieved_texts


# Load a text generation model from Hugging Face

try:
    from transformers import pipeline
    generator = pipeline(
        "text-generation",
        model="distilgpt2",  # Lightweight and runs locally
        tokenizer="distilgpt2",
        torch_dtype=torch.float32 # Specify dtype if needed for compatibility
    )
    print("Text generation pipeline loaded.")
except ImportError:
    print("transformers library not found. Skipping text generation pipeline.")
    generator = None
except Exception as e:
    print(f"Error loading text generation pipeline: {e}")
    generator = None

# Define RAG workflow
def rag_pipeline(query, top_k=3, max_length=250): # Increased max_length
    if generator is None:
        return "Text generation model not loaded. Cannot run RAG pipeline."

    # Retrieve context documents
    context_docs = retrieve_docs(query, top_k=top_k)
    context_text = "\n\n".join(context_docs) # Use double newline for separation

    # Build the prompt
    prompt = f"""Based on the following context, answer the question.

Context:
{context_text}

Question: {query}

Answer:"""

    # Generate a response
    try:
        response = generator(
            prompt,
            max_length=max_length,
            num_return_sequences=1,
            do_sample=True, # Enable sampling
            temperature=0.7, # Control creativity
            top_p=0.95,      # Control diversity
            pad_token_id=generator.tokenizer.eos_token_id # Handle padding
        )
        # Extract the generated text part after the prompt
        generated_text = response[0]["generated_text"]
        # Find where the answer likely starts
        answer_start = generated_text.find("Answer:")
        if answer_start != -1:
            generated_text = generated_text[answer_start + len("Answer:"):].strip()

        return generated_text

    except Exception as e:
        print(f"Error during text generation: {e}")
        return "Error generating response."


# Try a sample RAG query
query = "What are the latest trends in AI technology?"

# Only run RAG if generator is available
if generator:
    print("\nRunning RAG pipeline...")
    response = rag_pipeline(query)

    print("\n🔍 Query:", query)
    print("\n💡 RAG Response:\n", response)
else:
    print("\nSkipping RAG pipeline as text generation model could not be loaded.")
```
